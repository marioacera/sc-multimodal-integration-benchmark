```python
import umap

import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.metrics import adjusted_rand_score as ARI
from sklearn.metrics import adjusted_mutual_info_score as AMI

from scipy import sparse
from matplotlib import rcParams
rcParams['pdf.fonttype'] = 42 # enables correct plotting of text
rcParams['figure.figsize'] = (12,12)
import seaborn as sns

sc.settings.verbosity =0


import harmonypy as hm
# import scanorama
import matplotlib.pyplot as plt
import pandas as pd
import scanpy as sc
import scvi
from scib_metrics.nearest_neighbors import NeighborsResults, pynndescent
import scib_metrics

import pickle


import os
compute = True

import numpy as np
import pandas as pd
from scipy.sparse.csgraph import connected_components

import numpy as np
import pandas as pd
from scipy.sparse.csgraph import connected_components

# Ensure the directory for intermediate and final results exists
if not os.path.exists('CLISI_csv_PARTIAL'):
    os.makedirs('CLISI_csv_PARTIAL')
if not os.path.exists('CLISI_csv_FINAL'):
    os.makedirs('CLISI_csv_FINAL')
    
    
def binary_LISI(neigh, metadata):
    """
    Compute a binary version of the Local Inverse Simpson Index (LISI).

    :param neigh: A pynndescent object with the 'indices' attribute having shape (n_cells, n_neighbors)
    :param metadata: An array of shape (n_cells,) containing metadata labels for each cell
    :return: A numpy array of shape (n_cells,) with the binary LISI scores
    """
    n_cells = neigh.indices.shape[0]
    binary_lisi_scores = np.zeros(n_cells)

    for i in range(n_cells):
        center_label = metadata[i]
        neighbor_indices = neigh.indices[i]
        neighbor_labels = metadata[neighbor_indices]

        # Count how many neighbors have the same label as the center cell
        same_label_count = np.sum(neighbor_labels == center_label)
        
        # Compute the proportion of same label in the neighborhood
        binary_lisi_scores[i] = same_label_count / len(neighbor_indices)

    return binary_lisi_scores

def result_exists(mod, i, ann, n_size, ct):
    ct_sanitized = ct.replace('/', '_')  # Sanitize cell type name for use in filename
    filename = f'CLISI_csv_PARTIAL/scIB_{mod}_{i}_{ann}_BINARY_{n_size}_{ct_sanitized}.csv'
    return os.path.exists(filename), filename

def save_binary_lisi_scores(binary_lisi_scores, filename, valid_indices):
    temp_df = pd.DataFrame(binary_lisi_scores, index=valid_indices, columns=['binary_LISI'])
    print(temp_df.shape)

    temp_df.to_csv(filename)

```

```python
from scipy.stats import hmean
from scipy.spatial import distance



def compute_cell_type_centers(adata, cell_type_column='celltype', n_pcs=20):
    """
    Compute cell type centers using specified representation and PCs.
    """
    cell_type_centers = {}
    # print(adata,cell_type_column)
    for cell_type in adata.obs[cell_type_column].unique():
        cell_type_data = adata[adata.obs[cell_type_column] == cell_type]
        center = np.mean(cell_type_data.obsm['X_pca'][:, :n_pcs], axis=0)
        cell_type_centers[cell_type] = center
    return cell_type_centers

def cosine_distance(v1, v2):
    numerator = np.dot(np.mean(v1, axis = 0), np.mean(v2, axis = 0))
    denominator = np.linalg.norm(np.mean(v1, axis = 0)) * np.linalg.norm(np.mean(v2, axis = 0))
    return 1 - numerator / denominator

# Now, cov_matrix can be used to compute Mahalanobis distance

def euclidean_distance(v1, v2):
    """
    Calculate the Euclidean distance between two vectors.

    Parameters:
    - v1 (np.array): First vector.
    - v2 (np.array): Second vector.

    Returns:
    - float: The Euclidean distance.
    """
    # Calculate the difference vector
    diff = np.mean(v1, axis = 0) - np.mean(v2, axis = 0)
    
    # Compute the Euclidean distance
    dist = np.linalg.norm(diff)
    
    return dist

from scipy.stats import chi2
import numpy as np
import scipy.stats as stats



from sklearn.covariance import LedoitWolf


def unidirectional_mahalanobis_distance(v1, v2):
    """
    Calculate a directional Mahalanobis distance between one distribution and the centroid of the other,
    represented by the matrices v1 and v2, using their v1 covariance matrices.

    Parameters:
    - v1 (np.array): Matrix representing the first distribution, shape (n_obs1, n_features).
    - v2 (np.array): Matrix representing the second distribution, shape (n_obs2, n_features).
    
    Returns:
    - float: The composite Mahalanobis distance between the centroids of v1 and v2.
    """
    # Calculate the mean vectors of v1 and v2
    mean_v1 = np.mean(v1, axis=0)
    mean_v2 = np.mean(v2, axis=0)

    # Calculate the covariance matrices for v1 and v2
    # cov_v1 = np.cov(v1, rowvar=False)
    # cov_v2 = np.cov(v2, rowvar=False)
    lw = LedoitWolf()
    cov_v1 = lw.fit(v1).covariance_
    # cov_v2 = np.cov(v2, rowvar=False)

    # Calculate the inverse of the covariance matrices
    try:
        inv_cov_v1 = np.linalg.inv(cov_v1)
    except:
        return 0
    # inv_cov_v2 = np.linalg.inv(cov_v2)

    # Calculate the differences of the mean vectors
    delta = mean_v1 - mean_v2

    # Calculate Mahalanobis distance using each covariance matrix
    distance_v1 = np.sqrt(np.dot(np.dot(delta, inv_cov_v1), delta.T))
    # distance_v2 = np.sqrt(np.dot(np.dot(delta, inv_cov_v2), delta.T))

    # Composite distance can be the average of these two distances
    # composite_distance = (distance_v1 + distance_v2) / 2

    return distance_v1


def compute_distances_matrix(embedding, labels,n_cells=100, distance_function = cosine_distance, random_state =  np.random.randint(0,100)):
    unique_classes = np.unique(labels)
    distance_matrix = np.zeros((len(unique_classes), len(unique_classes)))
    for i, cls_i in enumerate(unique_classes):
        indices_i = np.where(labels == cls_i)[0]
        if len(indices_i) > n_cells:
            indices_i = sc.pp.subsample(indices_i, n_obs = n_cells, random_state = random_state)[0]
        for j, cls_j in enumerate(unique_classes):
            # print(cls_i,cls_j)
            if cls_i==cls_j:
                continue
            else:
                indices_j = np.where(labels == cls_j)[0]
                if len(indices_j) > n_cells:
                    indices_j = sc.pp.subsample(indices_j, n_obs = n_cells)[0]
                distance = distance_function(embedding[indices_i], embedding[indices_j])
            
                distance_matrix[i,j] = distance
                    
    return pd.DataFrame(distance_matrix, index = unique_classes, columns = unique_classes)

def symmetric_harmonic_mean(df):
    n_cols = df.shape[1]
    harmonic_means = pd.DataFrame(np.zeros((n_cols, n_cols)), columns=df.columns, index=df.columns)
    
    # Iterate over each pair of columns
    for i in range(n_cols):
        for j in range(i, n_cols):
            if i != j:
                try:
                    # Calculate the harmonic mean for each element pair-wise
                    val = hmean([df.iloc[i, j],df.iloc[i, j]])  # Adding a small number to avoid division by zero
                    harmonic_means.iloc[i, j] = val
                    harmonic_means.iloc[j, i] = val  # Ensure the matrix is symmetric
                except ZeroDivisionError:
                    # Handle division by zero if any values are exactly zero
                    harmonic_means.iloc[i, j] = np.nan
                    harmonic_means.iloc[j, i] = np.nan
    
    return harmonic_means
```

# SIMMETRIC MAHALANOBIS - POST REVIEWs - BOOSTRAPPING FOR STATISTICAL SIGNIFICANCE

```python
# distance_bootstrap_two_sample_obsreport.py
from __future__ import annotations
import os, itertools, traceback
from dataclasses import dataclass
from typing import Callable, Dict, List, Tuple, Optional

import numpy as np
import pandas as pd
import scanpy as sc
from sklearn.covariance import LedoitWolf
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt

# ============================ CONFIG ============================

CELLTYPE_LEVELS = [
    "Deepscore_HCA_l1_Clean",
    # "Deepscore_HCA_l3_Clean",
]

PRE_TECHS = {
    "pre_snRNA":   "../HORIZONTAL_RNA/objects/byTech/snRNA.h5ad",
    "pre_scRNA5p": "../HORIZONTAL_RNA/objects/byTech/scRNA5p.h5ad",
    "pre_scRNA":   "../HORIZONTAL_RNA/objects/byTech/scRNA.h5ad",
}
PRE_EMBED_KEY = "scVI"
PRE_NPCS: Optional[int] = None

POST_PATH = "../HORIZONTAL_RNA/objects/RNA_STWG_final.h5ad"
POST_EMBED_KEY = "scVI"
POST_BATCH_COL = "batch"
POST_BATCH_LEVELS = ["snRNA", "scRNA5p", "scRNA"]

# Distance + inference knobs (kept as in your approach)
DISTANCE_NAME = "mahalanobis"     # "mahalanobis" | "mahalanobis_unidir" | "euclidean" | "cosine"
B = 100                           # bootstrap draws for distance distributions
R = 500                           # resamples on the bootstrap distributions (fast)
ALPHA = 0.10                      # FDR level
CI_ALPHA = 0.10                   # 90% CI for Δ of means
ALTERNATIVE = "two-sided"         # "two-sided" | "less" | "greater" on Δ of means
FDR_METHOD = "fdr_tsbh"           # "fdr_bh" | "fdr_tsbh"

# Bootstrap behavior (cell-level), as in your snippet
BOOT_KIND = "nonparametric"       # "nonparametric" | "parametric" | "smoothed"
SMOOTH_JITTER_SD = 0.05           # for "smoothed"
MATCH_OBS_N = True                # resample to observed per-group sizes
DONOR_COL = None                  # e.g., "donor_id"
BOOT_STRATIFIED = False           # donor-stratified resampling if donor_col present

# Repro/IO
SEED = 42
OUTDIR = "results_two_sample_bootstrap"
CHKDIR = os.path.join(OUTDIR, "checkpoints")
LOGDIR = os.path.join(OUTDIR, "logs")

# Resume / redo policy
RESUME = True
FORCE_REDO = False
SKIP_EXISTING_GROUP_MATS = True
SKIP_EXISTING_COMPARISONS = True

# Optional fixed ordering
ORDER_L1 = ['IC','TAL','DCT','CNT','PC','PT','PEC','POD','EC','VSM/P','FIB','IMM']
ORDER_L3 = None

# ======================== Distances ==============================

def _mean_cov(X: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    mu = X.mean(axis=0)
    Sigma = LedoitWolf().fit(X).covariance_
    return mu, Sigma

def dist_mahalanobis_sym(X: np.ndarray, Y: np.ndarray) -> float:
    # EXACTLY your symmetric Mahalanobis: re-estimate both covariances each call
    muX, SX = _mean_cov(X); muY, SY = _mean_cov(Y)
    delta = (muX - muY).astype(float)
    S = 0.5*(SX + SY)
    try:
        sol = np.linalg.solve(S, delta)
        return float(np.sqrt(delta.dot(sol)))
    except np.linalg.LinAlgError:
        Sinv = np.linalg.pinv(S)
        return float(np.sqrt(delta.dot(Sinv).dot(delta)))

def dist_mahalanobis_unidir(X: np.ndarray, Y: np.ndarray) -> float:
    muX, SX = _mean_cov(X); muY = Y.mean(axis=0)
    delta = (muX - muY).astype(float)
    try:
        sol = np.linalg.solve(SX, delta)
        return float(np.sqrt(delta.dot(sol)))
    except np.linalg.LinAlgError:
        SXinv = np.linalg.pinv(SX)
        return float(np.sqrt(delta.dot(SXinv).dot(delta)))

def dist_euclidean(X: np.ndarray, Y: np.ndarray) -> float:
    return float(np.linalg.norm(X.mean(0) - Y.mean(0)))

def dist_cosine(X: np.ndarray, Y: np.ndarray) -> float:
    mX, mY = X.mean(0), Y.mean(0)
    num = float(np.dot(mX, mY))
    den = float(np.linalg.norm(mX) * np.linalg.norm(mY) + 1e-12)
    return 1.0 - num/den

DIST_FUNS: Dict[str, Callable[[np.ndarray, np.ndarray], float]] = {
    "mahalanobis": dist_mahalanobis_sym,
    "mahalanobis_unidir": dist_mahalanobis_unidir,
    "euclidean": dist_euclidean,
    "cosine": dist_cosine,
}

# ============================ Utils ==============================

rng_global = np.random.default_rng(SEED)
os.makedirs(OUTDIR, exist_ok=True)
os.makedirs(CHKDIR, exist_ok=True)
os.makedirs(LOGDIR, exist_ok=True)

def _get_latent(adata: sc.AnnData, key: str, n_pcs: Optional[int]) -> np.ndarray:
    Z = adata.obsm[key]
    if (key.lower().endswith("pca") or key == "X_pca") and n_pcs is not None:
        Z = Z[:, :n_pcs]
    return np.asarray(Z, dtype=float)

def _bootstrap_draw(Z: np.ndarray, rng: np.random.Generator, n: int,
                    kind: str = "nonparametric", smooth_sd: float = 0.0) -> np.ndarray:
    n0, p = Z.shape
    n = n0 if MATCH_OBS_N else min(n, n0)
    if kind == "nonparametric":
        idx = rng.integers(0, n0, size=n)
        return Z[idx]
    elif kind == "smoothed":
        idx = rng.integers(0, n0, size=n)
        jitter = rng.normal(0.0, smooth_sd, size=(n, p))
        return Z[idx] + jitter
    elif kind == "parametric":
        mu, Sigma = _mean_cov(Z)
        return rng.multivariate_normal(mean=mu, cov=Sigma, size=n)
    else:
        raise ValueError(f"Unknown BOOT_KIND={kind}")

# ======================== Data containers ========================

@dataclass
class Group:
    name: str
    adata: sc.AnnData
    embed_key: str
    n_pcs: Optional[int]
    mask: Optional[np.ndarray] = None

# =================== Core: distance distributions =================

def _cells_for_ct(ad: sc.AnnData, mask: np.ndarray, embed_key: str, n_pcs: Optional[int]) -> np.ndarray:
    return _get_latent(ad[mask], embed_key, n_pcs)

def observed_distance(ad: sc.AnnData, embed_key: str, n_pcs: Optional[int],
                      mask_i: np.ndarray, mask_j: np.ndarray,
                      distance: Callable[[np.ndarray, np.ndarray], float]) -> float:
    Xi = _cells_for_ct(ad, mask_i, embed_key, n_pcs)
    Xj = _cells_for_ct(ad, mask_j, embed_key, n_pcs)
    if len(Xi) < 3 or len(Xj) < 3: return np.nan
    return distance(Xi, Xj)  # ALL cells, no resampling

def bootstrap_distance_distribution(ad: sc.AnnData, embed_key: str, n_pcs: Optional[int],
                                    mask_i: np.ndarray, mask_j: np.ndarray,
                                    B: int, rng: np.random.Generator,
                                    distance: Callable[[np.ndarray, np.ndarray], float],
                                    donor_col: Optional[str], stratified: bool) -> np.ndarray:
    Xi = _cells_for_ct(ad, mask_i, embed_key, n_pcs)
    Xj = _cells_for_ct(ad, mask_j, embed_key, n_pcs)
    if len(Xi) < 3 or len(Xj) < 3: return np.full(B, np.nan, dtype=float)
    out = np.empty(B, dtype=float)

    if stratified and (donor_col is not None) and (donor_col in ad.obs.columns):
        di = ad.obs.loc[mask_i, donor_col].to_numpy()
        dj = ad.obs.loc[mask_j, donor_col].to_numpy()
        uniq_i, uniq_j = np.unique(di), np.unique(dj)
        stratified = (len(uniq_i) > 0 and len(uniq_j) > 0)
    else:
        stratified = False

    for b in range(B):
        if stratified:
            Zi_b = _resample_stratified(Xi, di, rng_global)
            Zj_b = _resample_stratified(Xj, dj, rng_global)
        else:
            Zi_b = _bootstrap_draw(Xi, rng, n=len(Xi), kind=BOOT_KIND, smooth_sd=SMOOTH_JITTER_SD)
            Zj_b = _bootstrap_draw(Xj, rng, n=len(Xj), kind=BOOT_KIND, smooth_sd=SMOOTH_JITTER_SD)
        out[b] = distance(Zi_b, Zj_b)  # re-estimate Mahalanobis each draw
    return out

def _resample_stratified(Z: np.ndarray, donors: np.ndarray, rng: np.random.Generator) -> np.ndarray:
    n0 = len(Z)
    uniq, counts = np.unique(donors, return_counts=True)
    if len(uniq) == 0:
        idx = rng.integers(0, n0, size=n0)
        return Z[idx]
    draw = rng.integers(0, len(uniq), size=len(uniq))
    idx_parts = []
    for d in draw:
        d_idx = np.where(donors == uniq[d])[0]
        n_d = counts[d]
        idx_parts.append(d_idx[rng.integers(0, len(d_idx), size=n_d)])
    idx = np.concatenate(idx_parts)
    if len(idx) > n0:
        idx = rng.choice(idx, size=n0, replace=False)
    elif len(idx) < n0 and len(idx) > 0:
        extra = rng.choice(idx, size=(n0 - len(idx)), replace=True)
        idx = np.concatenate([idx, extra])
    return Z[idx] if len(idx) else Z

# =================== Two-sample bootstrap test ===================

def two_sample_bootstrap_test(
    d_pre: np.ndarray, d_post: np.ndarray,
    R: int = 5000, alternative: str = "two-sided",
    ci_alpha: float = 0.10, rng: Optional[np.random.Generator] = None
) -> Tuple[float, Tuple[float,float], float]:
    """
    Compare the distributions of distances d_pre vs d_post:
    - Δ_boot_mean = mean(d_post) - mean(d_pre)
    - CI via bootstrap of the means (non-centered)
    - p-value via 'shifted' two-sample bootstrap (null: equal means)
    """
    if rng is None: rng = rng_global
    d_pre = d_pre[np.isfinite(d_pre)]; d_post = d_post[np.isfinite(d_post)]
    if len(d_pre) == 0 or len(d_post) == 0:
        return np.nan, (np.nan, np.nan), np.nan

    mu_pre, mu_post = float(np.mean(d_pre)), float(np.mean(d_post))
    delta_hat = mu_post - mu_pre

    nA, nB = len(d_pre), len(d_post)

    # CI for Δ via resampling the bootstrap replicates
    deltas_ci = np.empty(R, dtype=float)
    for r in range(R):
        mA = np.mean(d_pre[rng.integers(0, nA, size=nA)])
        mB = np.mean(d_post[rng.integers(0, nB, size=nB)])
        deltas_ci[r] = mB - mA
    ci_lo, ci_hi = np.percentile(deltas_ci, [100*ci_alpha/2, 100*(1-ci_alpha/2)])

    # Shifted null (equal means)
    m0 = 0.5 * (mu_pre + mu_post)
    pre0  = d_pre  - mu_pre + m0
    post0 = d_post - mu_post + m0

    deltas_null = np.empty(R, dtype=float)
    for r in range(R):
        mA = np.mean(pre0[rng.integers(0, nA, size=nA)])
        mB = np.mean(post0[rng.integers(0, nB, size=nB)])
        deltas_null[r] = mB - mA

    if alternative == "two-sided":
        p = (np.sum(np.abs(deltas_null) >= np.abs(delta_hat)) + 1) / (R + 1)
    elif alternative == "less":
        p = (np.sum(deltas_null <= delta_hat) + 1) / (R + 1)
    elif alternative == "greater":
        p = (np.sum(deltas_null >= delta_hat) + 1) / (R + 1)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    return float(delta_hat), (float(ci_lo), float(ci_hi)), float(p)

# ======================= Build groups / comps =====================

def build_groups_for_level(level_col: str) -> Tuple[List[Group], Dict[str, Group], List[Tuple[str,str]]]:
    groups: List[Group] = []
    for name, path in PRE_TECHS.items():
        ad = sc.read(path)
        groups.append(Group(name=name, adata=ad, embed_key=PRE_EMBED_KEY, n_pcs=PRE_NPCS, mask=None))
    ad_post = sc.read(POST_PATH)
    for b in POST_BATCH_LEVELS:
        mask = (ad_post.obs[POST_BATCH_COL].values == b)
        groups.append(Group(name=f"post_{b}", adata=ad_post, embed_key=POST_EMBED_KEY, n_pcs=None, mask=mask))
    gmap = {g.name: g for g in groups}
    comparisons = [(f"pre_{b}", f"post_{b}") for b in POST_BATCH_LEVELS if f"pre_{b}" in gmap and f"post_{b}" in gmap]
    return groups, gmap, comparisons

# ===================== Pre-mixture (weights) =====================

def mixture_weights(n_i: int, n_j: int) -> float:
    # harmonic mean of counts (pair support)
    if n_i <= 0 or n_j <= 0: return 0.0
    return 2.0 * n_i * n_j / (n_i + n_j)

def pre_mixture_bootstrap_and_observed(
    pre_groups: Dict[str, Group], level_col: str, i: str, j: str,
    B: int, rng: np.random.Generator, distance: Callable[[np.ndarray, np.ndarray], float]
) -> Tuple[np.ndarray, float]:
    """
    Bootstrap distance per pre tech (re-estimate Mahalanobis each draw),
    weight by harmonic-mean counts; return (d_pre_boot_mix, d_pre_obs_mix).
    """
    d_per_tech = []
    w_per_tech = []
    d_obs_tech = []
    for name, G in pre_groups.items():
        ad = G.adata
        base = np.ones(ad.n_obs, dtype=bool)
        mi = base & (ad.obs[level_col].astype(str).values == i)
        mj = base & (ad.obs[level_col].astype(str).values == j)
        Xi = _cells_for_ct(ad, mi, G.embed_key, G.n_pcs)
        Xj = _cells_for_ct(ad, mj, G.embed_key, G.n_pcs)
        if len(Xi) < 3 or len(Xj) < 3:
            continue

        rng_local = np.random.default_rng(rng.integers(0, 2**32-1))
        d_boot = bootstrap_distance_distribution(ad, G.embed_key, G.n_pcs, mi, mj,
                                                 B, rng_local, distance, DONOR_COL, BOOT_STRATIFIED)
        d_obs = distance(Xi, Xj)  # observed (all cells, no resampling)
        w = mixture_weights(len(Xi), len(Xj))
        if w <= 0: 
            continue
        d_per_tech.append(d_boot); d_obs_tech.append(d_obs); w_per_tech.append(w)

    if not d_per_tech:
        return np.full(B, np.nan), np.nan
    W = np.array(w_per_tech, dtype=float); W = W / W.sum()
    D = np.vstack(d_per_tech)             # (T, B)
    d_pre_boot_mix = (W[:, None] * D).sum(axis=0)
    d_pre_obs_mix  = float(np.dot(W, np.array(d_obs_tech, dtype=float)))
    return d_pre_boot_mix, d_pre_obs_mix

# ============================ Runner ==============================

def run_level(level_col: str):
    rng = np.random.default_rng(SEED)
    distance = DIST_FUNS[DISTANCE_NAME]

    # ----- pre vs post, technology-wise -----
    groups, gmap, comparisons = build_groups_for_level(level_col)
    order = ORDER_L1 if ("l1" in level_col.lower()) else ORDER_L3

    rows_prepost = []
    for pre_name, post_name in comparisons:
        Gpre, Gpost = gmap[pre_name], gmap[post_name]
        adA, adB = Gpre.adata, Gpost.adata
        baseA = Gpre.mask if Gpre.mask is not None else np.ones(adA.n_obs, dtype=bool)
        baseB = Gpost.mask if Gpost.mask is not None else np.ones(adB.n_obs, dtype=bool)

        ctsA = set(adA.obs.loc[baseA, level_col].dropna().astype(str).unique())
        ctsB = set(adB.obs.loc[baseB, level_col].dropna().astype(str).unique())
        cats = sorted(ctsA.intersection(ctsB))
        if order is not None:
            cats = [c for c in order if c in set(cats)]
        if len(cats) < 2:
            continue

        comp_id = f"{level_col}.{DISTANCE_NAME}.B{B}.R{R}.seed{SEED}.{post_name}_minus_{pre_name}"
        tmp_csv   = os.path.join(CHKDIR,  f"{comp_id}.pairs.tmp.csv")
        final_csv = os.path.join(OUTDIR, f"{comp_id}.pairs.csv")

        if RESUME and os.path.exists(final_csv) and SKIP_EXISTING_COMPARISONS and not FORCE_REDO:
            try:
                rows_prepost.append(pd.read_csv(final_csv))
                continue
            except Exception:
                pass

        header_written = os.path.exists(tmp_csv)
        def _append_tmp(rowdict):
            nonlocal header_written
            pd.DataFrame([rowdict]).to_csv(tmp_csv, mode="a", index=False, header=not header_written)
            header_written = True

        for i, j in itertools.combinations(cats, 2):
            try:
                miA = baseA & (adA.obs[level_col].values == i)
                mjA = baseA & (adA.obs[level_col].values == j)
                miB = baseB & (adB.obs[level_col].values == i)
                mjB = baseB & (adB.obs[level_col].values == j)

                rng_local = np.random.default_rng(rng.integers(0, 2**32-1))

                # bootstrap distance distributions (your approach)
                d_pre = bootstrap_distance_distribution(adA, Gpre.embed_key, Gpre.n_pcs, miA, mjA,
                                                        B, rng_local, distance, DONOR_COL, BOOT_STRATIFIED)
                d_post = bootstrap_distance_distribution(adB, Gpost.embed_key, Gpost.n_pcs, miB, mjB,
                                                         B, rng_local, distance, DONOR_COL, BOOT_STRATIFIED)
                if not (np.isfinite(d_pre).any() and np.isfinite(d_post).any()):
                    continue

                # inference from bootstrap distributions
                delta_boot_mean, (ci_lo, ci_hi), pval = two_sample_bootstrap_test(
                    d_pre, d_post, R=R, alternative=ALTERNATIVE, ci_alpha=CI_ALPHA, rng=rng_local
                )

                # --------- FINAL distances for reporting: one run with ALL cells ---------
                dist_pre_obs  = observed_distance(adA, Gpre.embed_key, Gpre.n_pcs, miA, mjA, distance)
                dist_post_obs = observed_distance(adB, Gpost.embed_key, Gpost.n_pcs, miB, mjB, distance)
                delta_obs = float(dist_post_obs - dist_pre_obs)

                row = {
                    "comparison": f"{post_name}_minus_{pre_name}",
                    "group_A": pre_name, "group_B": post_name,
                    "ct_i": i, "ct_j": j,
                    # observed (report these):
                    "dist_pre_obs":  float(dist_pre_obs),
                    "dist_post_obs": float(dist_post_obs),
                    "delta_obs":     delta_obs,
                    # bootstrap summaries (for transparency):
                    "mean_pre_boot": float(np.nanmean(d_pre)),
                    "mean_post_boot": float(np.nanmean(d_post)),
                    "delta_boot_mean": float(delta_boot_mean),
                    # inference:
                    "ci_lo": ci_lo, "ci_hi": ci_hi, "pval": pval,
                    "n_boot": B, "R_on_boot": R
                }
                _append_tmp(row)

            except Exception:
                with open(os.path.join(LOGDIR, f"{comp_id}.errors.log"), "a") as fh:
                    fh.write(f"PAIR {i}-{j}\n{traceback.format_exc()}\n")
                continue

        if os.path.exists(tmp_csv):
            df_pairs = pd.read_csv(tmp_csv)
            df_pairs["qval"] = multipletests(df_pairs["pval"].values, alpha=ALPHA, method=FDR_METHOD)[1]
            df_pairs["sig"] = df_pairs["qval"] < ALPHA
            df_pairs.sort_values(["qval","delta_obs"], ascending=[True,False]).to_csv(final_csv, index=False)
            os.remove(tmp_csv)
            rows_prepost.append(df_pairs)

    # ----- pre-mixture vs post-final (all technologies) -----
    # build pre_groups map first
    pre_groups: Dict[str, Group] = {}
    for name, path in PRE_TECHS.items():
        pre_groups[name] = Group(name=name, adata=sc.read(path), embed_key=PRE_EMBED_KEY, n_pcs=PRE_NPCS)

    ad_post_all = sc.read(POST_PATH)
    base_post_all = np.ones(ad_post_all.n_obs, dtype=bool)

    rows_premix = []
    for level_col in CELLTYPE_LEVELS:
        # shared cell types across pre techs and post
        cts_post = set(ad_post_all.obs.loc[base_post_all, level_col].dropna().astype(str).unique())
        cts_pre_all = set()
        for G in pre_groups.values():
            cts_pre_all |= set(G.adata.obs[level_col].dropna().astype(str).unique())
        cats = sorted(cts_pre_all.intersection(cts_post))
        order = ORDER_L1 if ("l1" in level_col.lower()) else ORDER_L3
        if order is not None:
            cats = [c for c in order if c in set(cats)]
        if len(cats) < 2:
            continue

        comp_id = f"{level_col}.{DISTANCE_NAME}.B{B}.R{R}.seed{SEED}.post_all_minus_pre_mixture"
        tmp_csv   = os.path.join(CHKDIR,  f"{comp_id}.pairs.tmp.csv")
        final_csv = os.path.join(OUTDIR, f"{comp_id}.pairs.csv")

        if RESUME and os.path.exists(final_csv) and SKIP_EXISTING_COMPARISONS and not FORCE_REDO:
            try:
                rows_premix.append(pd.read_csv(final_csv)); continue
            except Exception:
                pass

        header_written = os.path.exists(tmp_csv)
        def _append_tmp2(rowdict):
            nonlocal header_written
            pd.DataFrame([rowdict]).to_csv(tmp_csv, mode="a", index=False, header=not header_written)
            header_written = True

        for i, j in itertools.combinations(cats, 2):
            try:
                rng_local = np.random.default_rng(rng.integers(0, 2**32-1))
                # PRE mixture: bootstrap + observed (weighted)
                d_pre_boot_mix, d_pre_obs_mix = pre_mixture_bootstrap_and_observed(
                    pre_groups, level_col, i, j, B, rng_local, distance
                )
                if not np.isfinite(d_pre_boot_mix).any():
                    continue

                # POST final (all techs): bootstrap in final latent
                mi_post = base_post_all & (ad_post_all.obs[level_col].astype(str).values == i)
                mj_post = base_post_all & (ad_post_all.obs[level_col].astype(str).values == j)
                d_post = bootstrap_distance_distribution(ad_post_all, POST_EMBED_KEY, None, mi_post, mj_post,
                                                         B, rng_local, distance, DONOR_COL, BOOT_STRATIFIED)
                if not np.isfinite(d_post).any():
                    continue

                # inference from bootstrap distributions
                delta_boot_mean, (ci_lo, ci_hi), pval = two_sample_bootstrap_test(
                    d_pre_boot_mix, d_post, R=R, alternative=ALTERNATIVE, ci_alpha=CI_ALPHA, rng=rng_local
                )

                # FINAL distances (report): one pass with all cells
                dist_post_obs = observed_distance(ad_post_all, POST_EMBED_KEY, None, mi_post, mj_post, distance)
                delta_obs = float(dist_post_obs - d_pre_obs_mix)

                row = {
                    "comparison": "post_all_minus_pre_mixture",
                    "group_A": "pre_mixture", "group_B": "post_all",
                    "ct_i": i, "ct_j": j,
                    # observed (report):
                    "dist_pre_obs":  float(d_pre_obs_mix),   # weighted across pre techs
                    "dist_post_obs": float(dist_post_obs),
                    "delta_obs":     delta_obs,
                    # bootstrap transparency:
                    "mean_pre_boot":  float(np.nanmean(d_pre_boot_mix)),
                    "mean_post_boot": float(np.nanmean(d_post)),
                    "delta_boot_mean": float(delta_boot_mean),
                    # inference:
                    "ci_lo": ci_lo, "ci_hi": ci_hi, "pval": pval,
                    "n_boot": B, "R_on_boot": R
                }
                _append_tmp2(row)

            except Exception:
                with open(os.path.join(LOGDIR, f"{comp_id}.errors.log"), "a") as fh:
                    fh.write(f"PAIR {i}-{j}\n{traceback.format_exc()}\n")
                continue

        if os.path.exists(tmp_csv):
            df_pairs = pd.read_csv(tmp_csv)
            df_pairs["qval"] = multipletests(df_pairs["pval"].values, alpha=ALPHA, method=FDR_METHOD)[1]
            df_pairs["sig"] = df_pairs["qval"] < ALPHA
            df_pairs.sort_values(["qval","delta_obs"], ascending=[True,False]).to_csv(final_csv, index=False)
            os.remove(tmp_csv)
            rows_premix.append(df_pairs)

    # Write ALL compact table
    parts = []
    parts += rows_prepost
    parts += rows_premix
    if parts:
        res_all = pd.concat(parts, ignore_index=True)
        all_csv = os.path.join(OUTDIR, f"ALL.{DISTANCE_NAME}.B{B}.R{R}.seed{SEED}.pairs.csv")
        res_all.to_csv(all_csv, index=False)
        # summary for rebuttal
        summ = (res_all.assign(abs_delta=lambda d: d["delta_obs"].abs())
                        .groupby("comparison")
                        .agg(n_pairs=("pval","size"),
                             n_sig=("sig","sum"),
                             n_up=("delta_obs", lambda s: int((s>0).sum())),
                             n_dn=("delta_obs", lambda s: int((s<0).sum())),
                             med_abs_delta_sig=("abs_delta", lambda v: float(v[res_all.loc[v.index,"sig"]].median()) if int(res_all.loc[v.index,"sig"].sum()) else 0.0))
                        .reset_index())
        summ.to_csv(os.path.join(OUTDIR, f"SUMMARY.{DISTANCE_NAME}.csv"), index=False)
        print(summ)

# ============================= Main ==============================

# if __name__ == "__main__":
#     np.random.seed(SEED)
#     for level in CELLTYPE_LEVELS:
#         run_level(level)
```

```python
# make_all_plots.py
# Generate all requested plots from the *.pairs.csv produced by your distance pipeline.

import os, glob
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
from scipy.stats import chi2

# ------------------------------- CONFIG -------------------------------

PAIRS_DIR   = "results_two_sample_bootstrap"  # where your *.pairs.csv live
OUT_DIR     = "results_plots"                 # where to write figures
LEVEL_COL   = "Deepscore_HCA_l1_Clean"
DIST_NAME   = "mahalanobis"

# Which comparisons to render (filenames contain these)
# e.g., "post_snRNA_minus_pre_snRNA", "post_all_minus_pre_mixture"
COMPARISON_KEYS = [
    "post_snRNA_minus_pre_snRNA",
    "post_scRNA5p_minus_pre_scRNA5p",
    "post_scRNA_minus_pre_scRNA",
    "post_all_minus_pre_mixture",
]

# Optional subset/order of cell types for plotting (None -> all)
ORDER_L1 = ['IC','TAL','DCT','CNT','PC','PT','PEC','POD','EC','VSM/P','FIB','IMM']
PLOT_SUBSET = ['IC','TAL','DCT','CNT','PC','PT','PEC','POD']  # set None for all

# Heatmap knobs
ANNOTATE_NUMBERS = True
ANNOT_DECIMALS   = 2
GRID_LINEWIDTH   = 0.5
GRID_LINECOLOR   = "black"

# χ² overlap mapping (for Mahalanobis only)
LATENT_DIM = 30
BETA = LATENT_DIM / 3  # set to None to disable quadratic correction
MAKE_PRE_OVERLAP = False  # set True if you also want pre-mixture overlap plots

# ----------------------------------------------------------------------

os.makedirs(OUT_DIR, exist_ok=True)
sns.set_context("notebook")

# --------------------------- Utilities --------------------------------

def _best_fg_for_bg(val, norm, cmap):
    """Return black/white text depending on background luminance."""
    if not np.isfinite(val): return "black"
    r,g,b,_ = cmap(norm(val))
    def lin(c): return c/12.92 if c <= 0.04045 else ((c+0.055)/1.055)**2.4
    L = 0.2126*lin(r) + 0.7152*lin(g) + 0.0722*lin(b)
    return "black" if L > 0.5 else "white"

def _load_pairs_for_key(key: str) -> pd.DataFrame:
    pat = os.path.join(PAIRS_DIR, f"{LEVEL_COL}.{DIST_NAME}.B*.R*.seed*.{key}.pairs.csv")
    matches = sorted(glob.glob(pat), key=os.path.getmtime, reverse=True)
    if not matches:
        raise FileNotFoundError(f"No pairs CSV matching {pat}")
    return pd.read_csv(matches[0])

def _celltypes_from_pairs(df: pd.DataFrame, subset=None):
    cts_all = sorted(set(df["ct_i"]).union(set(df["ct_j"])))
    if subset is None:
        return cts_all
    return [c for c in subset if c in set(cts_all)]

# ----------------------- Matrix builders -------------------------------

def build_delta_matrix(df_pairs: pd.DataFrame, celltypes=None, value_col="delta_obs"):
    cts = _celltypes_from_pairs(df_pairs, celltypes)
    n = len(cts); idx = {c:i for i,c in enumerate(cts)}
    M = np.full((n,n), np.nan); S = np.zeros((n,n), dtype=bool)
    for r in df_pairs.itertuples():
        ci, cj = r.ct_i, r.ct_j
        if ci in idx and cj in idx:
            i,j = idx[ci], idx[cj]
            M[i,j] = M[j,i] = getattr(r, value_col)
            S[i,j] = S[j,i] = bool(getattr(r, "sig"))
    np.fill_diagonal(M, 0.0)
    return pd.DataFrame(M, index=cts, columns=cts), pd.DataFrame(S, index=cts, columns=cts)

def build_pre_post_half_matrix(df_pairs: pd.DataFrame, celltypes=None,
                               pre_col="dist_pre_obs", post_col="dist_post_obs"):
    cts = _celltypes_from_pairs(df_pairs, celltypes)
    n = len(cts); idx = {c:i for i,c in enumerate(cts)}
    M = np.full((n,n), np.nan); Star = np.zeros((n,n), dtype=bool)
    for r in df_pairs.itertuples():
        ci, cj = r.ct_i, r.ct_j
        if ci in idx and cj in idx:
            i,j = idx[ci], idx[cj]
            pre  = float(getattr(r, pre_col))
            post = float(getattr(r, post_col))
            sig  = bool(getattr(r, "sig"))
            dlt  = float(getattr(r, "delta_obs"))
            # upper triangle → post, lower → pre
            if i < j:
                M[i,j] = post
                if sig and dlt > 0:  # significant increase
                    Star[i,j] = True
            else:
                M[i,j] = pre
    np.fill_diagonal(M, 0.0)
    return pd.DataFrame(M, index=cts, columns=cts), pd.DataFrame(Star, index=cts, columns=cts)

def build_distance_matrix(df_pairs: pd.DataFrame, celltypes=None, value_col="dist_post_obs"):
    """Symmetric full matrix from pairs (e.g., for overlap mapping)."""
    cts = _celltypes_from_pairs(df_pairs, celltypes)
    n = len(cts); idx = {c:i for i,c in enumerate(cts)}
    M = np.full((n,n), np.nan)
    for r in df_pairs.itertuples():
        ci, cj = r.ct_i, r.ct_j
        if ci in idx and cj in idx:
            i,j = idx[ci], idx[cj]
            v = float(getattr(r, value_col))
            M[i,j] = M[j,i] = v
    np.fill_diagonal(M, 0.0)
    return pd.DataFrame(M, index=cts, columns=cts)

# --------------------------- Plotters ---------------------------------

def plot_delta_heatmap(D: pd.DataFrame, Sig: pd.DataFrame, out_png: str, title: str,
                       cmap="coolwarm", annotate=ANNOTATE_NUMBERS, decimals=ANNOT_DECIMALS):
    n = D.shape[0]
    v = np.abs(D.values[~np.eye(n, dtype=bool)])
    if v.size == 0:
        vlim = 1.0
    else:
        vlim = float(np.nanpercentile(v, 99))
    vmin, vmax = -vlim, vlim
    norm = mcolors.Normalize(vmin=vmin, vmax=vmax); cmap_obj = plt.get_cmap(cmap)

    fig, ax = plt.subplots(figsize=(max(4, 0.3*n), max(4, 0.3*n)))
    sns.heatmap(D, ax=ax, cmap=cmap, vmin=vmin, vmax=vmax, square=True, cbar=True,
                linewidths=GRID_LINEWIDTH, linecolor=GRID_LINECOLOR, annot=False)
    ax.set_title(title)
    cbar = ax.collections[0].colorbar; cbar.set_label("Δ distance (post − pre)")

    # star + numbers
    for i in range(n):
        for j in range(n):
            if i == j or not np.isfinite(D.iloc[i,j]): continue
            if annotate:
                ax.text(j, i, f"{D.iloc[i,j]:.{decimals}f}", ha="center", va="center",
                        fontsize=8, color=_best_fg_for_bg(D.iloc[i,j], norm, cmap_obj))
            if bool(Sig.iloc[i,j]) and D.iloc[i,j] > 0:
                # star in top-right corner
                ax.text(j+0.48, i-0.48, "*", ha="right", va="top",
                        fontsize=11, fontweight="bold",
                        color=_best_fg_for_bg(D.iloc[i,j], norm, cmap_obj))
    fig.tight_layout(); fig.savefig(out_png, dpi=250, bbox_inches="tight"); plt.close(fig)

def plot_pre_post_half(M: pd.DataFrame, Star: pd.DataFrame, out_png: str, title: str,
                       cmap="magma", annotate=ANNOTATE_NUMBERS, decimals=ANNOT_DECIMALS):
    n = M.shape[0]
    finite = M.values[~np.eye(n, dtype=bool)]
    vmax = float(np.nanpercentile(finite, 99)) if finite.size else 1.0
    vmin = 0.0
    norm = mcolors.Normalize(vmin=vmin, vmax=vmax); cmap_obj = plt.get_cmap(cmap)

    fig, ax = plt.subplots(figsize=(max(4, 0.3*n), max(4, 0.3*n)))
    sns.heatmap(M, ax=ax, cmap=cmap, vmin=vmin, vmax=vmax, square=True, cbar=True,
                linewidths=GRID_LINEWIDTH, linecolor=GRID_LINECOLOR, annot=False)
    ax.set_title(title)
    cbar = ax.collections[0].colorbar; cbar.set_label("distance (pre: lower; post: upper)")

    for i in range(n):
        for j in range(n):
            if i == j or not np.isfinite(M.iloc[i,j]): continue
            if annotate:
                ax.text(j, i, f"{M.iloc[i,j]:.{decimals}f}", ha="center", va="center",
                        fontsize=8, color=_best_fg_for_bg(M.iloc[i,j], norm, cmap_obj))
            if bool(Star.iloc[i,j]):
                ax.text(j+0.48, i-0.48, "*", ha="right", va="top",
                        fontsize=11, fontweight="bold",
                        color=_best_fg_for_bg(M.iloc[i,j], norm, cmap_obj))
    fig.tight_layout(); fig.savefig(out_png, dpi=250, bbox_inches="tight"); plt.close(fig)

def mahalanobis_to_overlap(D: pd.DataFrame, latent_dim: int, beta: float | None = None) -> pd.DataFrame:
    D2 = (D.values.astype(float))**2
    if beta is not None and beta > 0:
        D2 = D2 + beta * (1.0 - np.exp(-D2))
    O = chi2.sf(D2, df=latent_dim)
    np.fill_diagonal(O, 1.0)
    O = np.clip(O, 0.0, 1.0)
    return pd.DataFrame(O, index=D.index, columns=D.columns)

def overlap_summary_column(O: pd.DataFrame) -> pd.DataFrame:
    n = O.shape[0]
    if n <= 1:
        return pd.DataFrame(index=O.index, data={"summary": np.nan})
    s = 1.0 - (O.sum(axis=1) - 1.0) / (n - 1.0)
    return s.to_frame(name="summary")

def plot_overlap(O: pd.DataFrame, out_png: str, title: str, cmap="viridis"):
    n = O.shape[0]
    fig, ax = plt.subplots(figsize=(max(3, 0.3*n), max(3, 0.3*n)))
    sns.heatmap(O, ax=ax, cmap=cmap, vmin=0.0, vmax=1.0, square=True, cbar=True,
                linewidths=GRID_LINEWIDTH, linecolor=GRID_LINECOLOR, annot=False)
    ax.set_title(title)
    cbar = ax.collections[0].colorbar; cbar.set_label("overlap (χ² SF)")
    fig.tight_layout(); fig.savefig(out_png, dpi=250, bbox_inches="tight"); plt.close(fig)

def plot_summary_column_heatmap(summary_df: pd.DataFrame, out_png: str, title: str, cmap="mako"):
    fig_h = max(5, 0.35 * len(summary_df))
    fig, ax = plt.subplots(figsize=(3.2, fig_h))
    sns.heatmap(summary_df[::-1], ax=ax, cmap=cmap, vmin=0.0, vmax=1.0,
                cbar=True, linewidths=GRID_LINEWIDTH, linecolor=GRID_LINECOLOR, annot=False)
    ax.set_title(title); ax.set_xlabel(""); ax.set_ylabel("")
    cbar = ax.collections[0].colorbar; cbar.set_label("1 − average overlap")
    fig.tight_layout(); fig.savefig(out_png, dpi=250, bbox_inches="tight"); plt.close(fig)

# --- ADD: per-technology χ² overlaps for PRE and POST ---

def _chi2_overlap_per_tech_both(
    df_pairs: pd.DataFrame,
    level_col: str,
    which=("pre","post"),
    latent_dim_pre: int = LATENT_DIM,
    latent_dim_post: int = LATENT_DIM,
):
    """
    For each technology comparison (post_<tech>_minus_pre_<tech>), compute χ² overlaps
    for PRE and/or POST distances using seaborn heatmaps.
    """
    if df_pairs.empty:
        return

    for comp, dfc in df_pairs.groupby("comparison", sort=False):
        tech = comp.replace("post_", "").replace("_minus_pre_", "")

        if "pre" in which:
            out_prefix = os.path.join(OUTDIR, f"{level_col}.chi2_overlap.{tech}.pre")
            compute_and_plot_overlaps_from_pairs(
                df_pairs=dfc,
                value_col="dist_pre_obs",          # PRE distances (pre embedding)
                latent_dim=latent_dim_pre,
                beta=(latent_dim_pre/3.0),
                celltypes=PLOT_SUBSET,
                out_prefix=out_prefix,
                cmap="viridis",
            )

        if "post" in which:
            out_prefix = os.path.join(OUTDIR, f"{level_col}.chi2_overlap.{tech}.post")
            compute_and_plot_overlaps_from_pairs(
                df_pairs=dfc,
                value_col="dist_post_obs",         # POST distances (final embedding)
                latent_dim=latent_dim_post,
                beta=(latent_dim_post/3.0),
                celltypes=PLOT_SUBSET,
                out_prefix=out_prefix,
                cmap="viridis",
            )
# --- ADD: χ² overlaps for pre-mixture vs post-final, both sides ---

def _chi2_overlap_mix_vs_post_both(
    df_pairs: pd.DataFrame,
    level_col: str,
    which=("pre","post"),
    latent_dim_pre: int = LATENT_DIM,
    latent_dim_post: int = LATENT_DIM,
):
    """
    For the pre-mixture vs post-final table, compute χ² overlaps for:
      - PRE mixture distances (value_col='dist_pre_obs')
      - POST final distances (value_col='dist_post_obs')
    """
    if df_pairs.empty:
        return

    if "pre" in which:
        out_prefix = os.path.join(OUTDIR, f"{level_col}.chi2_overlap.pre_mixture")
        compute_and_plot_overlaps_from_pairs(
            df_pairs=df_pairs,
            value_col="dist_pre_obs",          # PRE mixture distances (weighted across techs)
            latent_dim=latent_dim_pre,
            beta=(latent_dim_pre/3.0),
            celltypes=PLOT_SUBSET,
            out_prefix=out_prefix,
            cmap="viridis",
        )

    if "post" in which:
        out_prefix = os.path.join(OUTDIR, f"{level_col}.chi2_overlap.post_final")
        compute_and_plot_overlaps_from_pairs(
            df_pairs=df_pairs,
            value_col="dist_post_obs",         # POST final distances (single final embedding)
            latent_dim=latent_dim_post,
            beta=(latent_dim_post/3.0),
            celltypes=PLOT_SUBSET,
            out_prefix=out_prefix,
            cmap="viridis",
        )


# ---------------------------- Driver ----------------------------------

def make_all_plots_for_comparison(key: str):
    df = _load_pairs_for_key(key)

    # Δ heatmap
    D, Sig = build_delta_matrix(df, celltypes=PLOT_SUBSET)
    plot_delta_heatmap(
        D, Sig,
        out_png=os.path.join(OUT_DIR, f"{key}.delta_heatmap.png"),
        title=f"Δ distance (post − pre) • {key}"
    )

    # Pre/Post half-matrix
    M, Star = build_pre_post_half_matrix(df, celltypes=PLOT_SUBSET,
                                         pre_col="dist_pre_obs", post_col="dist_post_obs")
    plot_pre_post_half(
        M, Star,
        out_png=os.path.join(OUT_DIR, f"{key}.pre_post_half.png"),
        title=f"Pre (↓) / Post (↑) distances • {key}"
    )

    # χ² overlap from POST distances only
    if "dist_post_obs" in df.columns:
        Dpost = build_distance_matrix(df, celltypes=PLOT_SUBSET, value_col="dist_post_obs")
        Opost = mahalanobis_to_overlap(Dpost, latent_dim=LATENT_DIM, beta=BETA)
        Opost.to_csv(os.path.join(OUT_DIR, f"{key}.post.overlap.csv"))
        plot_overlap(
            Opost,
            out_png=os.path.join(OUT_DIR, f"{key}.post.overlap_heatmap.png"),
            title=f"Overlap • POST only (χ² df={LATENT_DIM})"
        )
        summ_post = overlap_summary_column(Opost)
        summ_post.to_csv(os.path.join(OUT_DIR, f"{key}.post.overlap.summary.csv"))
        plot_summary_column_heatmap(
            summ_post,
            out_png=os.path.join(OUT_DIR, f"{key}.post.overlap.summary.columnheatmap.png"),
            title="1 − average overlap • POST"
        )

    # (Optional) χ² overlap from PRE distances (pre-mixture)
    if MAKE_PRE_OVERLAP and "dist_pre_obs" in df.columns:
        Dpre = build_distance_matrix(df, celltypes=PLOT_SUBSET, value_col="dist_pre_obs")
        Opre = mahalanobis_to_overlap(Dpre, latent_dim=LATENT_DIM, beta=BETA)
        Opre.to_csv(os.path.join(OUT_DIR, f"{key}.pre.overlap.csv"))
        plot_overlap(
            Opre,
            out_png=os.path.join(OUT_DIR, f"{key}.pre.overlap_heatmap.png"),
            title=f"Overlap • PRE (χ² df={LATENT_DIM})"
        )
        summ_pre = overlap_summary_column(Opre)
        summ_pre.to_csv(os.path.join(OUT_DIR, f"{key}.pre.overlap.summary.csv"))
        plot_summary_column_heatmap(
            summ_pre,
            out_png=os.path.join(OUT_DIR, f"{key}.pre.overlap.summary.columnheatmap.png"),
            title="1 − average overlap • PRE"
        )

# def main():
#     for key in COMPARISON_KEYS:
#         try:
#             make_all_plots_for_comparison(key)
#             print(f"[ok] Plotted: {key}")
#         except FileNotFoundError as e:
#             print(f"[skip] {key}: {e}")

# if __name__ == "__main__":
#     main()
```

```python
def main():
    np.random.seed(SEED)
    for level_col in CELLTYPE_LEVELS:
        # 1) Pre vs Post per technology
        df_prepost = compare_pre_vs_post_per_tech(level_col)
        _save_and_plot(df_prepost, tag="pre_vs_post_by_tech", level_col=level_col)
        _save_and_plot_per_tech(df_prepost, level_col=level_col)
        # χ² overlaps: BOTH pre and post per tech
        _chi2_overlap_per_tech_both(df_prepost, level_col, which=("pre","post"))

        # 2) Pre mixture vs Post final
        df_mix = compare_pre_mixture_vs_post_final(level_col)
        _save_and_plot(df_mix, tag="pre_mixture_vs_post_all", level_col=level_col)
        # χ² overlaps: BOTH pre-mixture and post-final
        _chi2_overlap_mix_vs_post_both(df_mix, level_col, which=("pre","post"))

        # 3) Aggregate ALL + SUMMARY (unchanged) ...
        frames = []
        if not df_prepost.empty: frames.append(df_prepost)
        if not df_mix.empty:     frames.append(df_mix)
        if frames:
            res_all = pd.concat(frames, ignore_index=True)
            res_all.to_csv(os.path.join(OUTDIR, f"ALL.{level_col}.pairs.csv"), index=False)
            summary = (
                res_all.assign(abs_delta=lambda d: d["delta_obs"].abs())
                       .groupby("comparison", as_index=False)
                       .agg(
                           n_pairs=("pval", "size"),
                           n_sig=("sig", "sum"),
                           n_up=("delta_obs", lambda s: int(((s > 0) & (res_all.loc[s.index, "sig"])).sum())),
                           n_dn=("delta_obs", lambda s: int(((s < 0) & (res_all.loc[s.index, "sig"])).sum())),
                           med_abs_delta_sig=("abs_delta", lambda v: float(v[res_all.loc[v.index, "sig"]].median()) if int(res_all.loc[v.index, "sig"].sum()) else 0.0),
                       )
            )
            summary.to_csv(os.path.join(OUTDIR, f"SUMMARY.{level_col}.csv"), index=False)
            print(summary)

if __name__ == "__main__":
    main()
```

```python
import matplotlib as mpl

import matplotlib as mpl

def custom_dotplot(
    adata,
    var_names,
    groupby,
    category_order=None,
    figsize=(12, 6),
    cmap="coolwarm",
    short_padding=0.1,
    long_padding=0.5,
    x_padding_interval=3,
    dot_min=0.05,
    dot_max=1.0,
    only_expressed=False,
    vmax=None,
    vmin=None
):
    """
    Custom dot plot with legends adjusted and uniform padding.

    Parameters
    ----------
    adata : AnnData
        Annotated data matrix.
    var_names : list
        List of genes to include in the plot.
    groupby : str
        Column in `adata.obs` to group cells by.
    category_order : list, optional
        Custom order for x-axis tick labels.
    figsize : tuple
        Size of the figure.
    cmap : str
        Colormap for the dots.
    short_padding : float
        Short padding applied between adjacent x-axis entries.
    long_padding : float
        Large padding applied every `x_padding_interval`.
    x_padding_interval : int
        Frequency of applying the long padding.
    dot_min : float
        Minimum size for dots.
    dot_max : float
        Maximum size for dots.
    vmax : float
        Maximum value for the color scale.
    vmin : float
        Minimum value for the color scale.

    Returns
    -------
    None
    """
    # Compute the percentage of cells expressing each gene and the mean expression
    categories = adata.obs[groupby].unique()
    if category_order:
        categories_order = category_order
    else:
        categories_order = sorted(categories)

    percentages = pd.DataFrame(
        {
            cat: (adata[adata.obs[groupby] == cat, var_names].X > 0).mean(axis=0).A1 * 100
            for cat in categories_order
        },
        index=var_names,
    )
    if only_expressed==True:
            
        mean_expression = pd.DataFrame(index=var_names)

        for cat in categories_order:
            mean_values = []
            for gene in var_names:
                # Extract expression data for the current group and gene
                gene_data = adata[adata.obs[groupby] == cat, gene].raw.X
                
                # Filter out cells with zero expression
                expressed_values = gene_data[gene_data > 0]
                
                # Compute the mean expression for non-zero values
                if expressed_values.size > 0.0:
                    mean_values.append(expressed_values.mean())
                else:
                    mean_values.append(0)  # Assign 0 if no cells express the gene
                
            # Add the mean expression values for the current category
            mean_expression[cat] = mean_values
    else:
        mean_expression = pd.DataFrame(
        {
            cat: adata[adata.obs[groupby] == cat, var_names].X.mean(axis=0).A1
            for cat in categories_order
        },
        index=var_names,
        )
    # Create a melted DataFrame for plotting
    data_for_plot = percentages.reset_index().melt(
        id_vars="index", var_name="Category", value_name="Percentage"
    )
    data_for_plot["Mean Expression"] = mean_expression.reset_index().melt(
        id_vars="index", var_name="Category", value_name="Expression"
    )["Expression"]

    # Adjust x positions with uniform padding
    x_positions = np.arange(len(categories_order))
    for i in range(1, len(x_positions)):
        if i % x_padding_interval == 0:
            x_positions[i:] = x_positions[i:] + long_padding
        else:
            x_positions[i:] = x_positions[i:] + short_padding

    # Map categories to x positions
    category_to_x = dict(zip(categories_order, x_positions))
    data_for_plot["x"] = data_for_plot["Category"].map(category_to_x)

    # Plot using seaborn scatterplot
    plt.figure(figsize=figsize)
    scatter_plot = sns.scatterplot(
        data=data_for_plot,
        x="x",
        y="index",
        size="Percentage",
        sizes=(dot_min, dot_max),
        hue="Mean Expression",
        palette=cmap,
        edgecolor="black",
        linewidth=0.5,
    )

    # Adjust plot aesthetics
    scatter_plot.set_xticks(x_positions)
    scatter_plot.set_xticklabels(categories_order, rotation=45, ha="right")
    scatter_plot.set_ylabel("Genes")
    scatter_plot.set_xlabel(groupby)

    # Add dot size legend outside the plot
    handles, labels = scatter_plot.get_legend_handles_labels()
    size_legend_handles = handles[:len(percentages.columns)]  # Size legend only
    size_legend_labels = labels[:len(percentages.columns)]

    scatter_plot.legend_.remove()  # Remove default legend
    plt.legend(
        handles=size_legend_handles,
        labels=size_legend_labels,
        title="Dot Size (%)",
        bbox_to_anchor=(1.05, 1),
        loc="upper left",
        borderaxespad=0,
    )

    # Add horizontal colorbar for Mean Expression on the right
    norm = mpl.colors.Normalize(vmin=0, vmax=vmax)
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])

    cbar = plt.colorbar(
        sm,
        ax=scatter_plot,
        orientation="horizontal",
        pad=0.17,
        aspect=50
    )
    cbar.set_label("Mean Expression")

    plt.tight_layout()
    # plt.show()
    return data_for_plot
data = custom_dotplot(
    adata=adata,
    var_names=ordered_genes, 
    groupby='celltype_batch', 
    category_order=celltype_batch_sorted,  # Custom order for x-ticks
    figsize=(14, 10),
    cmap="magma_r",
    short_padding=2.5,
    long_padding=4,
    x_padding_interval=3,
    dot_min=30,
    dot_max=230,
    vmax=5.0
)
plt.savefig('figures/3G_genes_DOTPLOT.png', dpi=600, bbox_inches='tight')
plt.savefig('figures/3G_genes_DOTPLOT.pdf', dpi=600, bbox_inches='tight')
```
